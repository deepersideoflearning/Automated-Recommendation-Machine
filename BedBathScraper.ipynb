```python
from selenium import webdriver
import os
import pandas as pd
import time
```

```python
driver = webdriver.Chrome(os.path.join(os.getcwd(), 'chromedriver'))

def getCategoryUrls(url):
    driver.get(url)
    driver.set_window_position(0, 0)
    driver.set_window_size(500, 768)
    category = driver.find_element_by_xpath('//*[@id="first"]/div[5]/ul/li[2]/a')
    category.click()
    urls = []
    for i in range(1, 10):
        try: 
            xpath = '//*[@id="first"]/div[4]/div[2]/div[' + str(i) + ']/div/div/a[1]'
            print(xpath)
            elem = driver.find_element_by_xpath(xpath)
            urls.append(elem.get_attribute("href"))
        except:
            break;
    driver.close()
    return urls


urls = getCategoryUrls('https://www.bedbathandbeyond.ca/store')
urls
```

```python
driver = webdriver.Chrome(os.path.join(os.getcwd(), 'chromedriver'))
def getProdUrl(urls):
    prodUrls = []
    for url in urls:
        driver.get(url)
        driver.set_window_position(0, 0)
        driver.set_window_size(500, 768)
#         time.sleep(0.5)
        elements = driver.find_elements_by_xpath("//*[@class='PrimaryLink_1RLwvm inline-block']")
        for elem in elements:
            prodUrls.append(elem.get_attribute("href"))
    driver.close()
    return prodUrls

prodUrls = getProdUrl(urls)
prodUrls
```

```python
driver = webdriver.Chrome(os.path.join(os.getcwd(), 'chromedriver'))
products = []
def getProductDetails(prodUrls):
    for url in prodUrls:
        driver.get(url)
        driver.set_window_position(0, 0)
        driver.set_window_size(500, 768)
        try:
            img = driver.find_element_by_xpath("//*[@class= 'HyperLink_6o9ywu ProductMediaCarouselStyle_XKpkhr']").get_attribute("href")
            name = driver.find_element_by_xpath("//*[@class='mb1 Title-inline_3PZ79c']").text
            price = driver.find_element_by_xpath("//*[@class='PDPPrice_3HTapM']").text
            details = driver.find_element_by_id('productDetails').text
            review = driver.find_element_by_xpath("//*[@class='Rating_3RTQ2U']").text
            details = details.replace('Details', '')
            details = details.replace('\n', '')
            products.append((str(img), name, price, details, review))
        except Exception as e:
            print(e)
    driver.close() 
    return products     

productDetails = getProductDetails(prodUrls[3:50])
```

```python
df = pd.DataFrame(productDetails)
df.columns = ['Image', 'Product', 'Price', 'Details', 'Reviews']
df.head()
```

```python
from pandas import ExcelWriter

writer = ExcelWriter('data.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```
